```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
# Define features (X) and target (y)
y = LDF["LapTime"]
X = LDF.drop(columns=["LapTime"])  # Keep all relevant features

# Identify categorical columns
categorical_cols = ['Driver']

# Convert categorical columns to 'category' typE

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42, shuffle=True)
for col in categorical_cols:
    X_train[col] = X_train[col].astype('category')
    X_test[col] = X_test[col].astype('category')

model = xgb.XGBRegressor(
    objective="reg:squarederror",
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    enable_categorical=True  # Enable support for categorical features
)

# Train the model
model.fit(X_train, y_train)

```

```python
y_pred = model.predict(X_test)
```

```python
# Calculate performance metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f} sec")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f} sec")
print(f"R² Score: {r2:.4f}")  # Closer to 1 = better model

print(f"Training Score: {model.score(X_train, y_train):.4f}")
print(f"Test Score: {model.score(X_test, y_test):.4f}")

```

```python
train_score = model.score(X_train, y_train)  # Training accuracy
test_score = model.score(X_test, y_test)    # Test accuracy

print(f"Training Accuracy: {train_score:.4f}")
print(f"Test Accuracy: {test_score:.4f}")
```
